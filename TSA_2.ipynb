```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Air Passengers dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
data = pd.read_csv(url, parse_dates=["Month"], index_col="Month")

# Set Seaborn style for better visuals
sns.set_style("whitegrid")

# 1. Plot the raw time series data
plt.figure(figsize=(10, 5))
plt.plot(data, color="blue", marker="o", linestyle="-")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.title("Monthly Air Passengers (1949-1960)")
plt.grid()
plt.show()

# 2. Histogram & Density Plot
plt.figure(figsize=(8, 5))
sns.histplot(data["Passengers"], kde=True, bins=20, color="blue")
plt.xlabel("Number of Passengers")
plt.ylabel("Frequency")
plt.title("Distribution of Passenger Count")
plt.grid()
plt.show()

# 3. Box Plot to Analyze Monthly Seasonality
data["Year"] = data.index.year
data["Month"] = data.index.month

plt.figure(figsize=(10, 5))
sns.boxplot(x=data["Month"], y=data["Passengers"], palette="coolwarm")
plt.xlabel("Month")
plt.ylabel("Number of Passengers")
plt.title("Monthly Seasonality in Air Passenger Data")
plt.grid()
plt.show()

# 4. Yearly Trend (Group by Year)
yearly_data = data.groupby("Year").mean()

plt.figure(figsize=(10, 5))
plt.plot(yearly_data.index, yearly_data["Passengers"], marker="o", linestyle="-", color="red")
plt.xlabel("Year")
plt.ylabel("Average Number of Passengers")
plt.title("Yearly Trend of Air Passengers (1949-1960)")
plt.grid()
plt.show()
```
